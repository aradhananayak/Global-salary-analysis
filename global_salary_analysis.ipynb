```python
pivot_df = pd.pivot_table(df, index='continent_name', values='average_salary')
pivot_df.plot(kind='bar', legend=False, figsize=(12,8))

plt.xlabel('Continent')
plt.ylabel('Average Salary')
plt.title('Average Salary by Continent')

plt.show()
```

```python
pivot_df = pd.pivot_table(df, index='continent_name', values='median_salary')
pivot_df.plot(kind='bar', legend=False, figsize=(12,8))


plt.xlabel('Continent')
plt.ylabel('Median Salary')
plt.title('Median Salary by Continent')


plt.show()
```

```python
pivot_df = pd.pivot_table(df, index='continent_name', values='highest_salary')
pivot_df.plot(kind='bar', legend=False, figsize=(12,8))

plt.xlabel('Continent')
plt.ylabel('Highest Salary')
plt.title('Highest Salary by Continent')

plt.show()
```

```python
pivot_df = pd.pivot_table(df, index='continent_name', values='lowest_salary')
pivot_df.plot(kind='bar', legend=False, figsize=(12,8))


plt.xlabel('Continent')
plt.ylabel('lowest Salary')
plt.title('lowest Salary by Continent')


plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df['median_salary'],
    df['average_salary'],
    c='b',
    marker='o',
    alpha=1,
    s=50,
    label='Data Points'
)

plt.title('Customized Scatter Plot of Median Salary vs. Average Salary')
plt.xlabel('Median Salary')
plt.ylabel('Average Salary')

plt.grid(True, linestyle='--', alpha=0.5)

cbar = plt.colorbar()
cbar.set_label('Colorbar Label')

plt.legend()

plt.show()
```
